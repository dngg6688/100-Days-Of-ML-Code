## 載入所需套件

```python
#數據、矩陣處理套件numpy
import numpy as np
#繪圖處理套件matplotlib
import matplotlib.pyplot as plt
#繪圖處理套件顯示中文matplotlib.font_manager
import matplotlib.font_manager as plt_font
#設定中文字體物件和字型檔案路徑
twfont1=plt_font.FontProperties(fname=folder+"字型/kaiu.ttf")
#產生亂數的套件
import random
```

## 定義雙變數函數f(x,y)=2(x-0.3)<sup>4</sup>+y<sup>4</sup>-4xy+1

```python
def F(x,y):
  return 2*(x-0.3)**4+y**4-4*x*y+1
```

## 定義雙變數函數的偏導函數

```python
def Fx(x,y):
  return 8*(x-0.3)**3*1-4*y
def Fy(x,y):
  return 4*y**3-4*x
```

## 設定梯度下降的參數

```python
#設定學習率，可以調大或調小，觀察梯度下降情況
ETA=0.01
#亂數取得梯度下降的起點
x_point=-2
y_point=2
#設定Step_x、Step_y、Step_z用來儲存梯度下降每一步的座標
Step_x=[]
Step_y=[]
Step_z=[]
```

## 用迴圈執行梯度下降

```python
for i in range(100):
    Step_x.append(x_point)
    Step_y.append(y_point)
    Step_z.append(F(x_point,y_point))
    #梯度下降的公式
    x_point=x_point-ETA*Fx(x_point,y_point)
    y_point=y_point-ETA*Fy(x_point,y_point)
```

## 劃出函數等高線圖及梯度下降路徑

等高線與梯度的方向垂直
且垂直是下降速度最快的

```python
#設定繪圖區大小
plt.figure(figsize=(10,10))
#繪圖區的標題，設定用中文字體twfont1，字體大小20
plt.title("等高線圖",fontproperties=twfont1,fontsize=20)
#設定橫軸和縱軸的標題
plt.xlabel("x",fontproperties=twfont1,fontsize=15)
plt.ylabel("y",fontproperties=twfont1,fontsize=15)
CS=plt.contour(x,y,F(x,y),200,cmap="cool")
plt.clabel(CS,inline=1,fontsize=10)#標等高線數值
plt.plot(Step_x,Step_y,"go")#繪製梯度下降路徑


```
